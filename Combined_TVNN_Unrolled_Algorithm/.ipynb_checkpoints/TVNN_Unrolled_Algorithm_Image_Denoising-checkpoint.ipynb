##### Define All Functions For TV-NN Algorithm & Train Model

```python
#soft_thresholding is a helper function that performs soft thresholding on an input array x with a threshold gamma. 
def soft_thresholding(x, gamma):
    retur
    n np.maximum(0, 1 - gamma / np.maximum(np.abs(x), 1E-10)) * x

#function that takes another proximal operator prox as input and returns a new proximal operator that operates on the dual variable
def dual_prox(prox):
    return lambda u, sigma: u - sigma * prox(u / sigma, 1 / sigma)

# maxiter_criterion is a function that checks if the iteration count is less than a specified maximum number of iterations (maxiter). 
#It is used as a stopping criterion in the optimisation process.
def maxiter_criterion(values, iteration, allvecs, fx, maxiter=100):
    return iteration < maxiter

#Context is a class used to encapsulate the optimisation context. 
#It contains information such as the stopping criterion, whether to return all iterations, and a callback function for evaluating the objective function
class Context(object):
    def __init__(self, criterion=maxiter_criterion,
                 full_output=False, retall=False, callback=None, **kwargs):
        self.criterion = criterion
        if 'maxiter' in kwargs:
            self.criterion = lambda v, i, a, f: maxiter_criterion(v, i, a, f, maxiter=kwargs['maxiter'])
        self.full_output = full_output
        self.retall = retall
        self.callback = callback

    def execute(self, values, step):
        allvecs = [values[0]]
        fx = []
        iteration = 0
        while self.criterion(values, iteration, allvecs, fx):
            values = step(*values)
            x = values[0]
            if self.full_output:
                pass
            if self.retall:
                allvecs.append(x)
            if self.callback:
                fx.append(self.callback(x))
            iteration += 1
        return self._output_helper(x, fx, allvecs)

    def _output_helper(self, x, fx, allvecs):
        if self.full_output:
            retlist = x, fx
            if self.retall:
                retlist += (allvecs,)
        else:
            retlist = x
            if self.retall:
                retlist = (x, allvecs)

        return retlist

defaultContext = Context()

def model(prox_fs, prox_g, K, x0, theta=1, sigma=None, tau=None, context=defaultContext):
    if isinstance(K, np.ndarray):
        op = lambda u: np.dot(K, u)
        op.T = lambda u: np.dot(K.T, u)
        return model(prox_fs, prox_g, op, x0, theta=theta,
                    sigma=sigma, tau=tau, context=context)
    if sigma is None or tau is None:  # Update the condition here
        L = operator_norm(
            lambda x: K.T(K(x)),
            np.random.randn(x0.shape[0], 1)
        )
        sigma = 10.0
        if sigma * L > 1e-10:
            tau = .9 / (sigma * L)
        else:
            tau = 0.0

    x = x0.copy()
    x1 = x0.copy()
    xold = x0.copy()
    y = K(x)

    def step(x, x1, xold, y):
        xold = x.copy()
        y = prox_fs(y + sigma * K(x1), sigma)
        x = prox_g(x - tau * K.T(y), tau)
        x1 = x + theta * (x - xold)
        return [x, x1, xold, y]

    return context.execute([x, x1, xold, y], step)

#operator_norm calculates the operator norm of a given linear operator. 
def operator_norm(linop, n=None, maxiter=30, check=False):
    if hasattr(linop, 'norm') and not check:
        return linop.norm
    if n is None:
        n = np.random.randn(linop.dim[1], 1)
    if np.size(n) == 1:
        u = np.random.randn(n, 1)
    else:
        u = n
    unorm = lin.norm(u)
    if unorm > 1e-10:
        u = u / unorm
    else:
        return 0
    e = []
    for i in range(maxiter):
        if hasattr(linop, 'T'):
            v = linop.T(linop(u))
        else:
            v = linop(u)
        e.append((u[:] * v[:]).sum())
        vnorm = lin.norm(v[:])
        if vnorm > 1e-10:
            u = v / vnorm
        else:
            return 0
    L = e[-1]
    return L

# Gradient and divergence with periodic boundaries
def gradient(x):
    g = np.zeros((x.shape[0], x.shape[1], 2))
    g[:, :, 0] = np.roll(x, -1, axis=0) - x
    g[:, :, 1] = np.roll(x, -1, axis=1) - x
    return g

def divergence(p):
    px = p[:, :, 0]
    py = p[:, :, 1]
    resx = px - np.roll(px, 1, axis=0)
    resy = py - np.roll(py, 1, axis=1)
    return -(resx + resy)

K = gradient
K.T = divergence
amp = lambda u: np.sqrt(np.sum(u ** 2, axis=2))
G = lambda x: 1 / 2 * np.linalg.norm(y - x, 'fro') ** 2

def amp(u):
    return tf.sqrt(tf.reduce_sum(u ** 2, axis=2))

def F(u, alpha):
    return alpha * tf.reduce_sum(amp(u))

normalize = lambda u: u / np.tile((np.maximum(amp(u), 1e-10))[:, :, np.newaxis],(1, 1, 2))
prox_f = lambda u, tau: np.tile(
    soft_thresholding(amp(u), alpha * tau)[:, :, np.newaxis],
    (1, 1, 2)) * normalize(u)
prox_fs = dual_prox(prox_f)
prox_g = lambda x, y, tau: (x + tau * y) / (1 + tau)

ctx = Context(full_output=True, maxiter=300)
ctx.callback = lambda x: G(x) + F(K(x))


def tv_loss(y_true, y_pred):
    alpha_pred = alpha_model(y_pred)
    alpha_pred_flat = tf.keras.layers.Flatten()(alpha_pred)

    y_pred_expanded = tf.expand_dims(y_pred, axis=0)

    tv_loss = tf.reduce_sum(tf.sqrt(tf.reduce_sum(tf.square(tf.image.image_gradients(y_pred_expanded)), axis=-1)))
    return tv_loss + F(y_pred, alpha_pred_flat)


def chambolle_pock_iteration(x, alpha, theta):
    y = K(x)
    return model(prox_fs, prox_g, K, x, theta=theta, sigma=alpha/10.0, tau=alpha/10.0, context=ctx)


#create_unrolled_model is a function that creates an unrolled neural network model for a specified number of iterations. 
#It defines a neural network with convolutional layers that mimic an iterative process.
def create_unrolled_model(num_iterations):
    input_shape = (64, 64, 1)
    input_layer = Input(shape=input_shape)
    x = Conv2D(32, (3, 3), padding='same')(input_layer)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)

    intermediate_outputs = []

    for _ in range(num_iterations):
        x = Conv2D(32, (3, 3), padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)
        x = Conv2D(32, (3, 3), padding='same')(x)
        x = BatchNormalization()(x)
        x = Activation('relu')(x)
        output = Conv2D(1, (3, 3), padding='same')(x)

        intermediate_outputs.append(output)

    if len(intermediate_outputs) > 1:
        x = Add()(intermediate_outputs)
    else:
        x = intermediate_outputs[0]

    final_model = tf.keras.Model(inputs=input_layer, outputs=x)

    x = GlobalAveragePooling2D()(x)

    alpha_estimator = Dense(1, activation='sigmoid', name='alpha_estimator')(x)

    alpha_model = tf.keras.Model(inputs=input_layer, outputs=alpha_estimator)

    return final_model, alpha_model

num_iterations=10
unrolled_model, alpha_model = create_unrolled_model(num_iterations)

unrolled_model.compile(optimizer='adam', loss=tv_loss)

#create_alpha_model is a neural network model used for estimating alpha values
def create_alpha_model(input_shape=(64, 64, 1)):
    model = Sequential()
    model.add(Input(shape=input_shape))
    model.add(Conv2D(32, (3, 3), padding='same', activation='relu'))
    model.add(Flatten())
    model.add(Dense(256, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))
    return model

alpha_model = create_alpha_model(input_shape=(64, 64, 1))
alpha_model.compile(optimizer='adam', loss='mse')

#fitting unrolled model with training data
unrolled_model.fit(train_noisy_images, train_clean_images, epochs=10, batch_size=16)
```

```python
#using the noisy images and clean images to calculate the alpha values for the training data

alpha_values = []

for noisy_image in train_noisy_images:
    noisy_image = noisy_image[np.newaxis, :, :, np.newaxis]
    alpha_value = alpha_model.predict(noisy_image)[0][0]
    alpha_values.append(alpha_value)

alpha_values = np.array(alpha_values)

alpha_model.fit(train_noisy_images, alpha_values, epochs=10, batch_size=16)
```

```python
#Run the Chambolle-Pock Algorithm using this estiamted alpha value to produce the denoised image

import numpy.linalg as lin


y = cv2.imread(noisy_image_test_path, cv2.IMREAD_GRAYSCALE)

# Regularisation parameter
alpha = estimated_test_alpha_value

# Gradient and divergence with periodic boundaries

def gradient(x):
    g = np.zeros((x.shape[0], x.shape[1], 2))
    g[:, :, 0] = np.roll(x, -1, axis=0) - x
    g[:, :, 1] = np.roll(x, -1, axis=1) - x
    return g


def divergence(p):
    px = p[:, :, 0]
    py = p[:, :, 1]
    resx = px - np.roll(px, 1, axis=0)
    resy = py - np.roll(py, 1, axis=1)
    return -(resx + resy)

# Minimisation of G(x) + F(K*x)
K = gradient
K.T = divergence
amp = lambda u: np.sqrt(np.sum(u ** 2, axis=2))
F = lambda u: alpha * np.sum(amp(u))
G = lambda x: 1 / 2 * lin.norm(y - x, 'fro') ** 2

# Proximity operators
normalize = lambda u: u / np.tile((np.maximum(amp(u), 1e-10))[:, :, np.newaxis],(1, 1, 2))
prox_f = lambda u, tau: np.tile(
    soft_thresholding(amp(u), alpha * tau)[:, :, np.newaxis],
    (1, 1, 2)) * normalize(u)
prox_fs = dual_prox(prox_f)
prox_g = lambda x, tau: (x + tau * y) / (1 + tau)


# context
ctx = Context(full_output=True, maxiter=300)
ctx.callback = lambda x: G(x) + F(K(x))

denoised_tvnn, cx = model(prox_fs, prox_g, K, y, context=ctx)
```
